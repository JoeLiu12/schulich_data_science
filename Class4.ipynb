```python
import numpy as np
import pandas as pd
import matplotlib as plt
import scipy.stats as stats
import seaborn as sns
```

```python
#Normal distribution
df = np.random.normal(loc=5000, scale = 10000, size = 1000)
```

```python
sns.histplot(df, kde=True)
```

```python
df.mean()
```

```python
df.std()
```

```python
#fit the distrubution to data
mu,std = stats.norm.fit(df)
```

```python
mu
```

```python
std
```

```python
#make predictions
stats.norm.cdf(6000,mu,std) - stats.norm.cdf(40000,mu,std)
```

```python
# Poisson distribution
arrivals = np.random.poisson(lam=10, size=1000)
```

```python
sns.histplot(arrivals,bins=50)
```

```python
time_to_purchase = np.random.gamma(shape=2, scale=1, size=1000)
```

```python
sns.histplot(time_to_purchase, kde=True)
```
